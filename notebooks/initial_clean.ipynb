```python
import pandas as pd 
import os
```

```python
df = pd.read_json(r'C:\Code Files\MonkeyTypeAutomation\data\raw\monkeytype.results.json')
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(columns=[col for col in ['uid', 'charStats', 'mode', 'mode2', 'incompleteTestSeconds', 
                                 'keyConsistency', 'chartData', 'name', 'keySpacingStats', 
                                 'keyDurationStats', 'afkDuration', 'difficulty', 'punctuation'] if col in df.columns], 
                                inplace=True)
```

```python
df = df.rename(columns={'_id': 'test_id',
                       'wpm': 'wpm',
                       'rawWpm': 'raw_wpm',
                       'charStats': 'char_stats',
                       'acc': 'accuracy',
                       'timestamp': 'time',
                       'restartCount': 'restart_count',
                       'testDuration': 'duration',
                       'consistency': 'consistency',
                       'isPb': 'personal_best'})
```

```python
df.head()
```

```python
df['restart_count'] = df['restart_count'].fillna(0)
df['restart_count'].head()
```

```python
df.head()
```

```python
df['personal_best'] = df['personal_best'].fillna(False)
df['personal_best'].head()
```

```python
df[df['personal_best'] == True].head()
```

```python
df['test_id'] = df['test_id'].apply(lambda x: x.get('$oid') if isinstance(x, dict) else x)
df['test_id'].head()
```

```python
df
```

```python
output_path = r'C:\Code Files\MonkeyTypeAutomation\data\processed\metrics.csv'
os.makedirs(os.path.dirname(output_path), exist_ok=True)
df.to_csv(output_path, index=False)
```
